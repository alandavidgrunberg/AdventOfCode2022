```python
with open('day4input.txt') as day4input:
    raw_input = day4input.read()

raw_input[:1000]
```

```python
pairs = raw_input.splitlines()
pairs[:10]
```

```python
# divide assigments to the 2 elves in each pair 

elf_a_assignments_strings = []
elf_b_assignments_strings = []

for pair in pairs:
    split_pair = pair.split(',')
    elf_a_assignments_strings.append(split_pair[0])
    elf_b_assignments_strings.append(split_pair[1])

print(elf_a_assignments_strings[0:5])
print(elf_b_assignments_strings[0:5])
```

```python
# convert assignments from strings to lists of ints 

elf_a_assignments = []
elf_b_assignments = []

for section_range in elf_a_assignments_strings:
    split = section_range.split('-')
    start = int(split[0])
    end = int(split[1]) + 1
    elf_a_assignments.append([int for int in range(start,end)])
    
for section_range in elf_b_assignments_strings:
    split = section_range.split('-')
    start = int(split[0])
    end = int(split[1]) + 1
    elf_b_assignments.append([int for int in range(start,end)])


print(elf_a_assignments[0:2])
```

```python
# iterate through pairs of assignment lists checking if either list fully contains the other,
# counting for how many pairs this is the case

full_containment_count = 0

for assignment_a,assignment_b in zip(elf_a_assignments, elf_b_assignments):
    
    if len(assignment_a) == len(assignment_b):
        if assignment_a == assignment_b:
            full_containment_count += 1 
    
    elif len(assignment_a) < len(assignment_b):
        a_in_b = 0
        for section in assignment_a:
            if section in assignment_b:
                a_in_b += 1
        if len(assignment_a) == a_in_b:
            full_containment_count += 1
    
    else:
        b_in_a = 0
        for section in assignment_b:
            if section in assignment_a:
                b_in_a += 1
        if len(assignment_b) == b_in_a:
            full_containment_count += 1

full_containment_count   
```

```python
# iterating through pairs of assignment lists checking if they overlap at all (contain any common ints),
# counting for how many pairs this is the case

overlap_count = 0

for assignment_a,assignment_b in zip(elf_a_assignments, elf_b_assignments):
    for section in assignment_a:
        if section in assignment_b:
            overlap_count += 1
            break

overlap_count
```
