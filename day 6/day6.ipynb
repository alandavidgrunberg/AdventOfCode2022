```python
with open('day6input.txt') as day6input:
    raw_input = day6input.read()

raw_input[:1000]
```

```python
# convert input to list of chars

input_list = [char for char in raw_input]
print(input_list[:10])
```

```python
# iterate through list of chars, viewing 4 chars at a time (first chars 1-4, then 2-5, then 3-6, etc.),
# and keeping count of how many characters you've viewed in total (+1 at each iteration)
# if all 4 chars are not unique, keep iterating and counting
# if all 4 chars are unique, stop iterating and return the count:
# this is how many chars had to be viewed before the start-of-packet marker (4 unique chars in a row) was found
 
count = 4 # count starts at 4 because we start the iteration by viewing 4 chars

for i in range(len(input_list)):
    window = input_list[i:i+4]
    window_set = set(window)
    if len(window_set) == 4:
        break
    else:
        count += 1

count
```

```python
# same process as above but this time viewing 14 chars at a time (first chars 1-14, then 2-15, etc.)
# keeping count of how many characters you've viewed in total (+1 at each iteration)
# if all 14 chars are not unique, keep iterating and counting
# if all 14 chars are unique, stop iterating and return the count:
# this is how many chars had to be viewed before the start-of-packet marker (14 unique chars in a row) was found


char_count = 14 # count starts at 14 because we start the iteration by viewing 14 chars

for i in range(len(input_list)):
    window = input_list[i:i+14]
    window_set = set(window)
    if len(window_set) == 14:
        break
    else:
        char_count += 1

char_count
```
