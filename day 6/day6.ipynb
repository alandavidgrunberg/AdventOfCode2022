```python
with open('day6input.txt') as day6input:
    raw_input = day6input.read()

raw_input
```

```python
input_list = [char for char in raw_input]
print(input_list[:10])
```

```python
len(input_list)
```

```python
char_count = 4

for i in range(0,4096):
    window = input_list[i:i+4]
    window_set = set(window)
    if len(window_set) == 4:
        break
    else:
        char_count += 1

char_count
```
