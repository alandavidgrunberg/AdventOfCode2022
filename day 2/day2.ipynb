```python
with open('day2input.txt') as day2input:
    raw_input = day2input.read()

raw_input[:1000]
```

```python
input_list = raw_input.splitlines()
input_list[:10]
```

```python
# 'A X' # rock vs rock 1 # tie 3 # total 4
# 'A Y' # rock vs paper 2 # win 6 # total 8
# 'A Z' # rock vs scissors 3 # lose 0 # total 3
# 'B X' # paper vs rock 1 # lose 0 # total 1
# 'B Y' # paper vs paper 2 # tie 3 # total 5
# 'B Z' # paper vs scissors 3 # win 6 # total 9
# 'C X' # scissors vs rock 1 # win 6 # total 7
# 'C Y' # scissors vs paper 2 # lose 0 # total 2
# 'C Z' # scissors vs scissors 3 # tie 3 # total 6


total_score = 0

for move in input_list:
    if move == 'A X':
        total_score += 4
    if move == 'A Y':
        total_score += 8
    if move == 'A Z':
        total_score += 3
    if move == 'B X':
        total_score += 1
    if move == 'B Y':
        total_score += 5
    if move == 'B Z':
        total_score += 9
    if move == 'C X':
        total_score += 7
    if move == 'C Y':
        total_score += 2
    if move == 'C Z':
        total_score += 6
```

```python
total_score
```
