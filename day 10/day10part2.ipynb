<b>TESTING ON EXAMPLE</b>

```python
with open('day10sample2input.txt') as day10sample:
    raw_input = day10sample.read()
```

```python
signals = raw_input.splitlines()

signals[:15]
```

```python
# same as function from part 1 of the puzzle, but this time keeping track of register value at every cycle

def program(signals):
   
    cycle = 0
    register = 1
    action_list = []
    
    registers_list = []
    
    for count, signal in enumerate(signals):
        
        cycle += 1
        registers_list.append(register)
        
        if signal == 'noop':
            action_list.append([1,0])
        else:
            value = int(signal.split()[1])
            action_list.append([2,value])
        
        if len(action_list) > 0:
            next_action = action_list[0]
            next_action[0] -= 1
            if next_action[0] == 0:
                register += next_action[1]
                action_list = action_list[1:]
        
    if count == len(signals)-1:
        
        while len(action_list) > 0:

            cycle += 1
            registers_list.append(register)
            
            next_action = action_list[0]
            next_action[0] -= 1
            if next_action[0] == 0:
                register += next_action[1]
                action_list = action_list[1:]
                 
    return registers_list
```

```python
# register value at every cycle

registers = program(signals)

print(registers[:40])
```

```python
# list from 0 to 40 representing the 40 pixel length of each line drawn

pixels = [p for p in range(40)]
print(pixels)
```

```python
# iterating through the pixels and first 40 registers:
#   if the pixel position == the register + or - 1,
#   pixel lights up '#'
#   otherwise it stays dark '.'

line1 = []

for cycle, register in zip(pixels, registers[:40]):
    if (cycle == register) or (cycle == register - 1) or (cycle == register + 1):
        line1.append("#")
    else:
        line1.append(".")
        
line1 = ''.join(line1)

print(line1)
```

```python
# same as above iterating through next 40 registers (line 2)

line2 = []

for cycle, register in zip(pixels, registers[40:80]):
    if (cycle == register) or (cycle == register - 1) or (cycle == register + 1):
        line2.append("#")
    else:
        line2.append(".")
        
line2 = ''.join(line2)

print(line2)
```

```python
# keep iterating through 40 registers at a time to draw each line

line3 = []

for cycle, register in zip(pixels, registers[80:120]):
    if (cycle == register) or (cycle == register - 1) or (cycle == register + 1):
        line3.append("#")
    else:
        line3.append(".")
        
line3 = ''.join(line3)

print(line3)
```

```python
line4 = []

for cycle, register in zip(pixels, registers[120:160]):
    if (cycle == register) or (cycle == register - 1) or (cycle == register + 1):
        line4.append("#")
    else:
        line4.append(".")
        
line4 = ''.join(line4)

print(line4)
```

```python
line5 = []

for cycle, register in zip(pixels, registers[160:200]):
    if (cycle == register) or (cycle == register - 1) or (cycle == register + 1):
        line5.append("#")
    else:
        line5.append(".")
        
line5 = ''.join(line5)

print(line5)
```
