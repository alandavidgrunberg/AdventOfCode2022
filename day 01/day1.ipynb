```python
with open('day1input.txt') as day1input:
    raw_input = day1input.read()

raw_input[:1000]
```

```python
input_list = raw_input.splitlines()
input_list[:20]
```

```python
# change cal counts from strings to ints and mark breaks between elves

cal_list = []

for item in input_list:
    if item == '':
        cal_list.append('break')
    else:
        cal_list.append(int(item))

cal_list[:20]
```

```python
# iterate through cal_list, summing up cal counts for each elves and saving them to sum_list

sum_list = []

cal_sum = 0
for item in cal_list:
    if item != 'break':
        cal_sum += item
    else:
        sum_list.append(cal_sum)
        cal_sum = 0 

sum_list[:10]
```

```python
# find elf carrying the most calories

max(sum_list)
```

```python
# sort sum_list from high to low

sum_list.sort(reverse=True)
sum_list[:10]
```

```python
# add up calories counted by top 3 elves 

top_3 = sum_list[0] + sum_list[1] + sum_list[2]
top_3
```
